#### Build a line graph

```python
import matplotlib.pyplot as plt
```

```python

x = [1,3,5,7,9]
y = [2,3,7,1,0]

#titulo
titulo = "Gr√°fico de linhas"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
# plt.legend()

#tipo de grafico
plt.plot(x,y)

#exibe grafico
plt.show()
#exibe grafico
plt.show()
```
