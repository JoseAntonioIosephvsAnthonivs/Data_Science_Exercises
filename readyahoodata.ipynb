```python
import pandas as pd
import numpy as np

import pandas_datareader as pdr
import datetime as dt

csv_file = 'AAPL.csv'
aapl = pd.read_csv(csv_file)
```

### Read Yahoo Data


```python
start = dt.datetime(2010,1,1)
aapl = pdr.get_data_yahoo("AAPL", start)
aapl.head()
```
