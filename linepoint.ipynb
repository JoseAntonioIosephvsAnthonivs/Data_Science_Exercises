#### Build a line point

```python
import matplotlib.pyplot as plt
```

```python
#parametros
x = [1,3,5,7,9]
y = [2,3,7,1,0]

#titulo
titulo = "Grafico de dispersão"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)


#tipo de grafico
plt.scatter(x,y)
plt.plot(x,y)

#execução
plt.show()
```
